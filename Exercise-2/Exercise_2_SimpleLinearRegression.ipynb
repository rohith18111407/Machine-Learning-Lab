```python
class SimpleLinearRegration:
    global a0
    global a1
    def fit(leabel,target):
        xy = (target * leabel).sum()
        x_square_sum = (leabel ** 2).sum()

        mean_x = leabel.mean()
        mean_y = target.mean()

        a1 = (len(df) * xy - sum(leabel) * sum(target)) / (len(df) * x_square_sum - (sum(leabel) ** 2))
        a0 = mean_y - a1 * mean_x

    def predict(val):
        return a0 + a1 * val

    def line(test_leabel):
        return a0 + a1 * test_leabel

x = df['TV'][0]
y = df['Sales'][0]
print("val =",x," Result = ", SimpleLinearRegration.predict(x)," Target value = ", y)
x = df['TV'][1]
y = df['Sales'][1]
print("val =",x," Result = ", SimpleLinearRegration.predict(x)," Target value = ", y)

```

```python
pos = 49

x = df['TV'][pos]
y = df['Sales'][pos]
print("val =",x," Result = ", SimpleLinearRegration.predict(x)," Target value = ", y)

plt.scatter(df['TV'], df['Sales'], color='blue', label='Data Points')
predicted_TV = SimpleLinearRegration.line(df['TV'])
plt.plot(df['TV'], predicted_TV, color='red', label='Regression Line')
plt.scatter(x, y, color='black', label='Test Data')
plt.xlabel('Sales')
plt.ylabel('TV')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```
